# EXPLORANDO LA LECTURA Y ESCRITURA DE ARCHIVOS


````python
'''EN ESTAS LINEAS DE CODIGO SE MUESTRA COMO ALMACENAR LA INFORMACION QUE GENERAMOS EN EL CODIGO QUE PIDE
RESUMENES DE EMAILS CON GROQ. EL OBJETIVO ES ALMACENAR LOS RESUMENES EN UN ARCHIVO DE TEXTO'''

emails = [
    "Hola, confirmo mi asistencia a la reunión programada para mañana a las 10:00 AM.",
    "Buenos días, tengo problemas para acceder a mi cuenta y necesito ayuda.",
    "Gracias por su interés. Adjunto la cotización solicitada para su revisión.",
    "Le recordamos que la fecha límite para entregar los documentos es este viernes.",
    "Su pedido ha sido enviado y llegará en los próximos tres días hábiles.",
]
````

````python
#Todo este bloque se basa en el realizado con AI de gemini para resumir los correos
#este bloque se modifica para trabajar con groq
#importante ver que tambien se modico la parte de importar la IA, instalar groq, variables (client, API, completion y model)
# en lugar de contents se copia y se pega el message completo del objeto que nos dio el simulador de groq
import os
from google.colab import userdata
from groq import Groq
os.environ['GROQ_API_KEY'] = userdata.get('GROQ_API_KEY')
client = Groq()

def resumir_emails(emails):
  lista_resumenes = [] #primera modificacion, creacion de una lista vacia para almacenar los resumenes
  for numero, email in enumerate(emails):
    #la siguiente parte es donde se carga el modelo de groq completion....
    # en la parte de mensaje se pone el formato que pide groq para hacer request, la cual es una lista de diccionarios
     respuesta = client.chat.completions.create(
        model="openai/gpt-oss-120b",
        messages=[
                {
                      "role": "system",
                      "content": f"""dame un resumen del email o correo, la respuesta debe ser unicamente de una linea.
                      El email es el siguiente {email}""" #en esta parte es donde cambiamos la palabra
                }
          ]
        )
     texto = respuesta.choices[0].message.content #asi es como se saca solo la respuesta cuando se pide algo a groq, a diferencia de gemini que se pone respuesta.text
     lista_resumenes.append(f"E-mail Num {numero+1}: {texto}") #segunda modificacion, se anade a la lista los strings de los resumenes
  return lista_resumenes
````

````python
#se llama la funcion a ver si funciona
#se debe de imprimir la lista de resumenes que se retorno
lista_emails_resumidos = resumir_emails(emails)
````

````python
lista_emails_resumidos
````

````python
with open("lista-resumenes.txt","w", encoding="utf-8") as archivo: #esto es
#   1              2             3           4              5
#1 funcion para guardar archivos
#2 primer argumento de la funcion que es el nombre de archivo
#3 accion w = write o a guardar
#4 indica el mapa de caracteres, encoding="utf-8" es el equivalente al idioma espanol, tiene todos los caracteres del alfabeto
#5 se indica el alias con el que se quiere guardar o el tipo de formato

  #archivo.writelines(resumir_emails(emails)) ... con esta linea de codigo se guarda la lista, en una sola linea
  for resumen in lista_emails_resumidos: #itera por cada uno de los items de la lista retornada por la funcion "resumir_emails(emails)" que son los string formateado
    archivo.write(resumen+"\n") #otro metodo para escribir en archivos y cada item se separa por
````

````python
import os
from google.colab import userdata
from groq import Groq
os.environ['GROQ_API_KEY'] = userdata.get('GROQ_API_KEY')
client = Groq()
````

````python

def respuesta_preguntas_lista(lista_preguntas): #recordar que el argumento aqui puede o no tener el mismo nombre que la variable que introduciremos a la funcion
  lista_respuestas = ["Pregunta, Respuesta"]
  for pregunta in lista_lectura_preguntas:
    respuesta = client.chat.completions.create(
        model="openai/gpt-oss-120b",
        messages=[
                {
                      "role": "system",
                      "content": f"""Dame una respuesta de una sola linea
                      a la pregunta que te hago {pregunta}""" #en esta parte es donde cambiamos la palabra
                }
          ]
        )
    texto = respuesta.choices[0].message.content
    #print(f"Respuesta {numero+1}: {texto}") asi es para obtenerla sin forma de lista
    lista_respuestas.append(f'"{pregunta}", {texto}') #es necesario poner la respuesta asi para que pandas identifique que pregunta
    #y texto van en la misma fila pero en columanas separadas
    #las comillas varias es porque Las comillas le dicen a pandas
    #"todo lo que está dentro de estas comillas es un solo valor, ignora las comas internas, si no con cada coma
    #dentro de las respuestas se romperia el formato y lo interpretaria como otra columna":
  return lista_respuestas
````

MISMO DESAFIO PERO REALIZADO EN CLASE

Lo que cambia en el codigo de la clase es que en lugar de guardar la respuesta de la IA en una lista, lo hace en un diccionario. Aparte el dataframe para pandas se llama de manera diferente

````python
def respuesta_preguntas_dicionario(lista_preguntas): #recordar que el argumento aqui puede o no tener el mismo nombre que la variable que introduciremos a la funcion
  lista_diccionarios_respuestas = []
  for pregunta in lista_lectura_preguntas:
    respuesta = client.chat.completions.create(
        model="openai/gpt-oss-120b",
        messages=[
                {
                      "role": "system",
                      "content": f"""Dame una respuesta de una sola linea
                      a la pregunta que te hago {pregunta}""" #en esta parte es donde cambiamos la palabra
                }
          ]
        )
    texto = respuesta.choices[0].message.content
    #print(f"Respuesta {numero+1}: {texto}") asi es para obtenerla sin forma de lista
    lista_diccionarios_respuestas.append({"pregunta": pregunta, "respuesta": texto})

  return lista_diccionarios_respuestas
````

````python
lista_diccionarios_P_R_final = respuesta_preguntas_dicionario(lista_lectura_preguntas)
````

````python
import os
from google.colab import userdata
from groq import Groq
os.environ['GROQ_API_KEY'] = userdata.get('GROQ_API_KEY')
client = Groq()
````

````python
#Asignar las reviews a lista vacia
lista_de_anaisis = []
````

````python
for numero, review in enumerate(columna_de_reviews):
  #recordar que numero es un contador de cada iteracion
  #reviw es el valor de la iteracion actual o valor de cada fila
  #enumarate es la funcion que permite usar la enumeracion de cada iteracion
  #el argumenoto en el parentesis es la columna sobre la que vamos a iterar previamente filtrada
    respuesta = client.chat.completions.create(
      model="openai/gpt-oss-120b",
      messages=[
          {

                      "role": "system",
                      "content": f"""Debes analizar la reseña que yo te envio y debes de darme como respuesta un análisis
                      de sentimiento. Debes responder con: Positiva, Negativa o Neutra.
                      Ejemplos:

                      me encantó el producto -> Positiva
                      El producto no es nada especial -> Neutra
                      No me gustó el producto -> Negativa

                      Debes respondr solamente una de las tres opciones (positivo, negativo o neutro)
                      no agregues texto o explicaciones

                      La reseña es la siguiente {review}"""


                }

          ]
        )
    texto = respuesta.choices[0].message.content #asi es como se saca solo la respuesta cuando se pide algo a groq, a diferencia de gemini que se pone respuesta.text

    print(f"Review {numero+1}: {review} -> Sentimientos: {texto}")

    lista_de_anaisis.append(texto) #con cada iteracion se pone en la lista vacia
    #la respuesta de la IA la cual es: negativo, neutro o positivo solamente
````
